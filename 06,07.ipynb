### melt 함수

> id_vars : 위치를 그대로 유지할 열의 이름을 지정

> value_vars : 행으로 위치를 변경할 열의 이름을 지정

> var_name : value_vars로 위치를 변경한 열의 이름을 지정

> value_name : var_name으로 위치를 변경한 열의 데이털르 저장한 열의 이름을 지정

### pivot_table

> 행과 열의 위치를 다시 바꿔 정리

> index 인자에는 위치를 그대로 유지할 열 이름을 지정

> columns 인자에는 피벗할 열 이름을 지정

> values 인자에는 새로운 열의 데이터가 될 열 이름을 지정

```python
wether_tidy = weather_melt.pivot_table(index=['id','year','month','day'], columns = 'element' , values = 'temp')
```
